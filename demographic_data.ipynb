```python
import pandas as pd


column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num',
    'marital-status', 'occupation', 'relationship', 'race', 'sex',
    'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'salary'
]


df = pd.read_csv('adult.data', header=None, names=column_names)

def calculate_demographic_data(print_data=True):
    result = {}

    # 1. Count of each race
    result['race_count'] = df['race'].value_counts()

    # 2. Average age of men
    result['average_age_men'] = round(df[df['sex'] == 'Male']['age'].mean(), 1)

    # 3. Percentage with Bachelor's degree
    result['percentage_bachelors'] = round(
        (df['education'] == 'Bachelors').mean() * 100, 1
    )

    # 4. Advanced education vs high earners
    higher_edu = df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])
    lower_edu = ~higher_edu

    result['higher_education_rich'] = round(
        (df[higher_edu]['salary'] == '>50K').mean() * 100, 1
    )

    result['lower_education_rich'] = round(
        (df[lower_edu]['salary'] == '>50K').mean() * 100, 1
    )

    # 5. Min hours and % of rich among them
    min_hours = df['hours-per-week'].min()
    result['min_work_hours'] = min_hours

    num_min_workers = df[df['hours-per-week'] == min_hours]
    result['rich_percentage'] = round(
        (num_min_workers['salary'] == '>50K').mean() * 100, 1
    )

    # 6. Country with highest % rich
    country_income = df[df['salary'] == '>50K']['native-country'].value_counts() / df['native-country'].value_counts()
    highest_country = country_income.idxmax()
    result['highest_earning_country'] = highest_country
    result['highest_earning_country_percentage'] = round(country_income.max() * 100, 1)

    # 7. Top occupation in India among high earners
    india_50k = df[(df['native-country'] == 'India') & (df['salary'] == '>50K')]
    if not india_50k.empty:
        result['top_IN_occupation'] = india_50k['occupation'].value_counts().idxmax()
    else:
        result['top_IN_occupation'] = 'No high earners in India found'


    if print_data:
        for k, v in result.items():
            print(f"{k}: {v}\n")

    return result
```

```python
calculate_demographic_data()
```
